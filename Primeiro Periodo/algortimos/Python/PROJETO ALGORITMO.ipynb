```python
def linha(tam=104):
    return '-'*tam


def cabeçalho (txt):
    print (linha())
    print (txt.center(104))
    print (linha())

    
def menu(lista):
    #cabeçalho('MENU')
    c=1
    for item in lista:
        print(f'{c}-{item}')
        c +=1
    print (linha())



def alternativas(lista):
    print (linha())
    c=1
    for item in lista:
        print(f'{c}-{item}')
        c +=1
    print (linha())
    

def Pesquisa_e_Registro(Df, nome, num):
    while num != 1 and num != 2:
        
        num = input('digite um valor valido')
        
    while num == 1:
        
        nome = input("Nome do livro:")
        display (Df[Df["Título"].str.contains(pat=nome,case=False)])
        alternativas (["Pesquisar Novamente", "Fazer registro"])
        num = int(input("Opção:"))
            
    if num == 2:
        
        registro_df = pd.read_excel ("registro9.xlsx")
        aluno = input("Nome do Aluno:")
        titulo = input("Título do livro:")
        while True:
            try:
                dia_registro = pd.to_datetime(input("Dia do Registro:"), format='%d/%m/%Y').date() #o format so formata como vai ser a entrada e nao como ela vai ser mostrada
                dia_entrega = pd.to_datetime(input ("Dia da Entrega:"), format='%d/%m/%Y').date()
            except ValueError:
                print ("Digite a data corretamente.Formato: dd/mm/aaaa")
                continue 
            else:
                break
                
        dia_entregue = ""
        registro_df.loc[len(registro_df)] = [aluno , titulo, dia_registro,dia_entrega, dia_entregue]
        registro_df ['Dia do registro'] = pd.to_datetime(registro_df['Dia do registro']) #transforma p/ data
        registro_df ['Dia da entrega'] = pd.to_datetime(registro_df['Dia da entrega'])
        registro_df ['Dia entregue'] = pd.to_datetime(registro_df['Dia entregue'])
        registro_df.to_excel('registro9.xlsx', encoding='utf',index=False)
        Df.loc[Df['Título'] == titulo, 'Quantidade'] = Df.loc[Df['Título'] == titulo, 'Quantidade']-1
        print ("Registro realizado com sucesso!")
        alternativas (['Voltar'])
        categoria = int(input("Opção:"))
        
        return Df

Df = None
    
#----------------------------------------------------------LOGIN-------------------------------------------------------------

    
    
#------------------------------------------------------MENU------------------------------------------------------------------

cabeçalho('MENU')
menu (['Categorias', 'Reservas', 'Entregas','Cadastrar livro', 'Sair'])
opcao = int(input("Opção:"))
    
#-------------------------------------------------- CATEGORIAS---------------------------------------------------------------
while opcao != 5:
    if opcao ==  1:

        cabeçalho('CATEGORIAS')
        menu (['Matemática', 'Física', 'Filosofia', 'Suspense', 'Romance',"Voltar"])
        categoria = int(input("Opção:"))


        if categoria == 1:
            
            cabeçalho('MATEMÁTICA')
            matematica_df = pd.read_excel ("livrosMatematica1.xlsx")
            display(matematica_df)
            nome_mat = input("Nome do livro:")
            display (matematica_df[matematica_df["Título"].str.contains(pat=nome_mat,case=False)])
            alternativas (["Pesquisar Novamente", "Fazer registro","Voltar"])
            alternativa = int(input("Opção:"))

            Pesquisa_e_Registro(matematica_df,nome_mat,alternativa )
            matematica_df.to_excel('livrosMatematica1.xlsx', encoding='utf',index=False)    


        elif categoria == 2:
            
            cabeçalho('FÍSICA')
            fisica_df = pd.read_excel ("livrosFisica.xlsx")
            display(fisica_df)
            nome_fis = input("Nome do livro:")
            display (fisica_df[fisica_df["Título"].str.contains(pat=nome_fis,case=False)])
            alternativas (["Pesquisar Novamente", "Fazer registro","Voltar"])
            alternativa = int(input("Opção:"))

            Pesquisa_e_Registro(fisica_df,nome_fis,alternativa )
            fisica_df.to_excel('livrosMatematica.xlsx', encoding='utf',index=False)    
           

        elif categoria == 3:
            
            cabeçalho('FILOSOFIA')
            filosofia_df = pd.read_excel ("livrosFilosofia.xlsx")
            display(filosofia_df)
            nome_filo = input("Nome do livro:")
            display (filosofia_df[filosofia_df["Título"].str.contains(pat=nome_filo,case=False)])
            alternativas (["Pesquisar Novamente", "Fazer registro","Voltar"])
            alternativa = int(input("Opção:"))

            Pesquisa_e_Registro(filosofia_df,nome_filo,alternativa )
            filosofia_df.to_excel('livrosMatematica.xlsx', encoding='utf',index=False)    
           

        elif categoria == 4:
            
            cabeçalho('SUSPENSE')
            suspense_df = pd.read_excel ("livrosSuspense.xlsx")
            display(suspense_df)
            nome_sus = input("Nome do livro:")
            display (suspense_df[suspense_df["Título"].str.contains(pat=nome_sus,case=False)])
            alternativas (["Pesquisar Novamente", "Fazer registro","Voltar"])
            alternativa = int(input("Opção:"))

            Pesquisa_e_Registro(suspense_df,nome_sus,alternativa )
            suspense_df.to_excel('livrosMatematica.xlsx', encoding='utf',index=False)    
            

        elif categoria == 5:
            
            cabeçalho('ROMANCE')
            romance_df = pd.read_excel ("livrosRomance.xlsx")
            display(romance_df)
            nome_romance = input("Nome do livro:")
            display (romance_df[romance_df["Título"].str.contains(pat=nome_romance,case=False)])
            alternativas (["Pesquisar Novamente", "Fazer registro","Voltar"])
            alternativa = int(input("Opção:"))

            Pesquisa_e_Registro(romance_df,nome_romance,alternativa )
            romance_df.to_excel('livrosMatematica.xlsx', encoding='utf',index=False)    
            
        
        elif categoria == 6:
            
            cabeçalho('MENU')
            menu (['Categorias', 'Reservas','Entregas','Cadastrar livro', 'Sair'])
            opcao = int(input("Opção:"))




    
#--------------------------------------------------------------RESERVAS--------------------------------------------------------      

    elif opcao == 2:

        cabeçalho('RESERVAS')
        menu (["Fazer Reserva","Visualizar Reservas","Retirar reserva","Voltar"])
        reserva = int(input("Opção:"))
        reserva_df = pd.read_excel ("reserva.xlsx")

        if reserva == 1: 
    
            cabeçalho('FAZER RESERVA')
            aluno = input("Nome do aluno:")
            
            if aluno in reserva_df.values:    #verificação do aluno
                print ("Não pode fazer a reserva")
                alternativas (['Voltar'])
                reserva = int(input("Opção:"))

            else:         
                
                titulo = input("Título do livro:")
                data = input("Dia da reserva:")
                reserva_df.loc[len(reserva_df)] = [aluno , titulo, data]
                reserva_df.to_excel('reserva.xlsx', encoding='utf',index=False)
                display(reserva_df)
                alternativas (['Voltar'])
                reserva = int(input("Opção:"))

        elif reserva == 2: #visualizar reserva
            
            cabeçalho('VISUALIZAR RESERVAS')
            display (reserva_df)
            alternativas (['Voltar'])
            reserva = int(input("Opção:"))

        elif reserva == 3: #retirar reserva
            
            cabeçalho('RETIRAR RESERVA')
            display (reserva_df)
            index = int(input("Indíce:"))
            reserva_df.drop ([index], axis=0, inplace=True )
            display (reserva_df)
            alternativas (['Voltar'])
            reserva = int(input("Opção:"))
        
        elif reserva == 4: #volta p/o menu
            
            cabeçalho('MENU')
            menu (['Categorias', 'Reservas', 'Entregas','Cadastrar livro', 'Sair'])
            opcao = int(input("Opção:"))
 


    #--------------------------------------------------------- ENTREGAS-----------------------------------------------------------
    elif opcao == 3:

        registro_df = pd.read_excel ("registro9.xlsx")
        alternativas (["Adicionar dia que o livro foi entregue","Multas","Voltar"])


        while True:
            try:
                alternativa = int(input("Opção:"))
                if alternativa == 1:
                    display (registro_df.iloc[:,0:5])
                    index = int(input("Index:"))
                    dia_entregue = pd.to_datetime (input("Dia Entregue:"), format='%d/%m/%Y').date() #o format so formata como vai ser a entrada e nao como ela vai ser mostrada
                    alternativas (['Voltar'])
                    alternativa = int(input("Opção:"))
                    
                if alternativa == 2:

                    a= (registro_df['Dia entregue'] - registro_df['Dia da entrega']).dt.days #transformar os dias p/int
                    b = (a)*0.50
                    registro_df['Preço'] =b #criar coluna com o preço b
                    registro_df['Preço'] = registro_df['Preço'].clip(lower=0) #transformar os preços negativos em 0
                    display (registro_df)
                    alternativas (['Voltar'])
                    alternativa = int(input("Opção:"))
                
                elif alternativa == 3:
                    
                    cabeçalho('MENU')
                    menu (['Categorias', 'Reservas', 'Entregas','Cadastrar livro', 'Sair'])
                    opcao = int(input("Opção:"))          

            except ValueError:
                print ("Valor incorreto, digite novamente")
                continue
            else:
                break


        registro_df.iat [index,4] =  dia_entregue
        registro_df ['Dia entregue'] = pd.to_datetime(registro_df['Dia entregue']) # tem que repetir aqui pq se nao fica com o tempo
        registro_df.to_excel('registro9.xlsx', encoding='utf',index=False)



    #-----------------------opcao para fazer a adicao de livros novos nas tabelas de cada categoria------------------------------
    elif opcao == 4: 

        cabeçalho ("ADICIONAR LIVROS")
        menu (['Matemática', 'Física', 'Filosofia', 'Suspense', 'Romance','Voltar'])
        categoria = int(input("Opção:"))

        if categoria == 1:
            cabeçalho ("MATEMÁTICA")
            matematica_df = pd.read_excel ("livrosMatematica1.xlsx")
            titulo = input('Título:')
            autor = input('Autor: ')
            idioma = input('Idioma')
            qtd = int(input('Quantidade:'))
            matematica_df.loc[len(matematica_df)] = [titulo , autor,"Matemática", idioma, qtd]
            matematica_df.to_excel('livrosMatematica1.xlsx', encoding='utf',index=False)
            display(matematica_df)
            alternativas (['Voltar'])
            categoria = int(input("Opção:"))

        elif categoria == 2:
            cabeçalho ("FÍSICA")
            fisica_df = pd.read_excel ("livrosFisica.xlsx")
            titulo = input('Título:')
            autor = input('Autor: ')
            idioma = input('Idioma')
            qtd = int(input('Quantidade:'))
            fisica_df.loc[len(fisica_df)] = [titulo , autor,"Física", idioma, qtd]
            display(fisica_df)
            alternativas (['Voltar'])
            categoria = int(input("Opção:"))

        elif categoria == 3:
            cabeçalho ("FILOSOFIA")
            filosofia_df = pd.read_excel ("livrosFilosofia.xlsx")
            titulo = input('Título:')
            autor = input('Autor: ')
            idioma = input('Idioma')
            qtd = int(input('Quantidade:'))
            filosofia_df.loc[len(filosofia_df)] = [titulo , autor,"Filosofia", idioma, qtd]
            display(filosofia_df)
            alternativas (['Voltar'])
            categoria = int(input("Opção:"))

        elif categoria == 4:
            cabeçalho ("SUSPENSE")
            suspense_df = pd.read_excel ("livrosSuspense.xlsx")
            titulo = input('Título:')
            autor = input('Autor: ')
            idioma = input('Idioma')
            qtd = int(input('Quantidade:'))

            suspense_df.loc[len(suspense_df)] = [titulo , autor,"Suspense", idioma, qtd]
            suspense_df.to_excel('livrosSuspense.xlsx', encoding='utf',index=False) #adiciona no arquivo
            display(suspense_df) #so mostra
            alternativas (['Voltar'])
            categoria = int(input("Opção:"))

        elif categoria == 5:
            cabeçalho ("ROMANCE")
            romance_df = pd.read_excel ("livrosRomance.xlsx")
            titulo = input('Título:')
            autor = input('Autor: ')
            idioma = input('Idioma')
            qtd = int(input('Quantidade:'))
            romance_df.loc[len(romance_df)] = [titulo , autor,"Romance", idioma, qtd]
            display(romance_df)
            alternativas (['Voltar'])
            categoria = int(input("Opção:"))
            
        elif categoria == 6:
            cabeçalho('MENU')
            menu (['Categorias', 'Reservas', 'Entregas','Cadastrar livro', 'Sair'])
            opcao = int(input("Opção:"))

        #ate aqui    
    #elif opcao == 6:
        #print("Volta para a tela de login")

    else:
        print("Resultado inválido, tente novamente!")
        cabeçalho('MENU')
        menu (['Categorias', 'Reservas', 'Entregas','Cadastrar livro', 'Sair'])
        opcao = int(input("Opção:"))
        
print("Obrigado, volte sempre!")
```
